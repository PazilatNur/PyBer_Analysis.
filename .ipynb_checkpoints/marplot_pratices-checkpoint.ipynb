```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#usying matplot funtion to create chart
plt.plot(x_axis,y_axis)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Line chart")
# plt.xlim("Jan","June")
# OR
#plt.xlim(0,5)
#plt.ylim(0, 50)
# Add a grid.
# plt.grid()
# Add the legend.
# plt.legend()
plt.show
```

```python
# Set the x-axis to a list of strings for each month.
xb_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
yb_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.bar(xb_axis, yb_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
#To create a horizontal chart, we use the plt.barh() function:
plt.barh(yb_axis, xb_axis)
```

```python
plt.barh(xb_axis, yb_axis)
plt.gca().invert_yaxis()
#invert the y-axis to have the months in ascending order, 
#use the gca() method. The gca() method means "get current axes." 
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(xb_axis, yb_axis, color="black", label='New York')
```

```python
fig, ax = plt.subplots()
ax.barh(y_axis, x_axis)
```
